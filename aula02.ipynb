##PASSO 1 - Buscar automaticamente os dados das ações

```python
import yfinance
```

```python
codigo = input("Digite o código da ação desejada  ")
dados = yfinance.Ticker(codigo).history("6mo")
fechamento = dados.Close
fechamento.plot()
```

##PASSO 2 - GERAR AS ANÁLISES DE FORMA AUTOMÁTICA

*   Análise dos últimos seis meses
*   Cotação maxíma
*   Cotação atual





```python
cotacao_maxima = fechamento.max()
cotacao_minima = fechamento.min()
cotacao_atual  = fechamento[-1]
print(cotacao_atual)
```

##PASSO 3 - Enviar um email para o nosso gestor
* abrir uma nova aba no navegador (crtl+t)
* digitar o endereço do gmail e dar um enter
* clicar no botão escrever
* colocar o email do destinatário (tab)
* colocar o assunto(tab)
* corpo do email
* clicar no botão enviar





```python
!pip install pyautogui
```

```python
import pyautogui
import pyperclip
```

```python
pyautogui.PAUSE = 2

#abrir uma nova aba no navegador (ctrl + t)
pyautogui.hotkey("ctrl", "t")

#digitar o endereço do gmail e dar um enter
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

#clicar no botão escrever
pyautogui.click(x=2029, y=226)

#colocar o email do destinatário (tab)
pyperclip.copy("gestor@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

#colocar o assunto (tab)
pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

#corpo do email
mensagem = f"""
Prezado gestor,

Seguem as análises diárias dos últimos seis meses da ação {codigo}:

Cotação máxima: R${round(cotacao_maxima, 2)}
Cotação mínima: R${round(cotacao_minima, 2)}
Cotação atual:  R${round(cotacao_atual, 2)}

Qualquer dúvida, fico à disposição!
"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

#clicar no botão enviar
pyautogui.click(x=3120, y=977)
```
